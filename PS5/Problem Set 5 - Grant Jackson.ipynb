#### (a) - OLS Estimation with Time Dummies

```python
import statsmodels.formula.api as smf

# Creating new dataset with year dummies
data_dummies = pd.get_dummies(data, columns=['year'], drop_first=True)

data_dummies
```

```python
data_dummies = data_dummies.dropna()
data_dummies.info()
```

```python
# Removing the '01-01 00:00:00' from the dummie column names
data_dummies.columns = [str(col).replace(' 00:00:00', '').replace('-01-01', '') if 'year' in str(col) else col for col in data_dummies.columns]

data_dummies.columns
```

```python
# Defining OLS model with time dummies
model_ols = smf.ols('y ~ lt + k + year_1983 + year_1984 + year_1985 + year_1986 + year_1987 + year_1988 + year_1989 + year_1990',
                    data=data_dummies
).fit()

print(model_ols.summary())
```

``Interpretation: Coefficient for lt is 0.671, meaning a 1% increase in temporary labor is associated with a 0.671% increase in output. Coefficient for k is 0.284, meaning a 1% increase in new capital is associated with a 0.284% increase in output.``

#### (b) - Test the Null Hypothesis lambda_L = 1

```python
from scipy.stats import norm

# Extracting values
lambda_L = model_ols.params['lt']
lambda_L_se = model_ols.bse['lt']

# Calculating t-stat and p-value
t_stat_lambda_L = (lambda_L - 1) / lambda_L_se
p_value_lambda_L = 2 * (1 - norm.cdf(abs(t_stat_lambda_L)))

print('Test for lambda L = 1:')
print('lambda_L =', lambda_L)
print('SE =', lambda_L_se)
print('t-stat =', t_stat_lambda_L)
print('p_value =', p_value_lambda_L)
```

``Interpretation: Reject the null hypothesis that lambda_L = 1. Suggests that temporary labor is significantly less productive than permanent labor.``

#### (c) - Test the Null Hypothesis lambda_K = 1

```python
# Extracting values
lambda_K = model_ols.params['k']
lambda_K_se = model_ols.bse['k']

# Calculating t-stat and p-value
t_stat_lambda_K = (lambda_K - 1) / lambda_K_se
p_value_lambda_K = 2 * (1 - norm.cdf(abs(t_stat_lambda_K)))

print('Test for lambda K = 1:')
print('lambda_K =', lambda_K)
print('SE =', lambda_K_se)
print('t-stat =', t_stat_lambda_K)
print('p_value =', p_value_lambda_K)
```

``Interpretation: Reject null hypothesis that lambda_K = 1. Suggests that new capital is significantly less productive than old capital.``

#### (d) - Test the Null Hypothesis aL + aK = 1

```python
# Extracting values
alpha_L = model_ols.params['lt']
alpha_K = model_ols.params['k']
alpha_sum = alpha_L + alpha_K

# Calculating SE, t-stat, and p-value for the sum
alpha_sum_se = (model_ols.cov_params().loc['lt', 'lt'] + model_ols.cov_params().loc['k', 'k'] + 2 * model_ols.cov_params().loc['lt', 'k']) ** 0.5
t_stat_alpha_sum = (alpha_sum - 1) / alpha_sum_se
p_value_alpha_sum = 2 * (1 - norm.cdf(abs(t_stat_alpha_sum)))

print('Test for aL + aK = 1:')
print('alpha_L=', alpha_L)
print('alpha_K', alpha_K)
print('t-stat =', t_stat_alpha_sum)
print('p_value =', p_value_alpha_sum)
```

``Interpretation: Reject null hypothesis that aL + aK = 1. Combined productivity of labor and capital less than 1 is significant evidence of decreasing returns to scale.``

#### (a) - Fixed Effects Estimation with Time Dummies

```python
# Creating new dataset and setting the index for panel data structure 
year_dummies = pd.get_dummies(data['year'], prefix='year', drop_first=True)

year_dummies.columns = [str(col).replace(' 00:00:00', '').replace('-01-01', '') if 'year' in str(col) else col for col in year_dummies.columns]

fe_data = pd.concat([data, year_dummies], axis=1)

fe_data.set_index(['id', 'year'], inplace=True)

fe_data
```

```python
fe_data = fe_data.dropna()
fe_data.info()
```

```python
# Defining the model using fixed effects estimator with time dummies
from linearmodels.panel import PanelOLS

fe_model = PanelOLS.from_formula(
    'y ~ lt + k + year_1983 + year_1984 + year_1985 + year_1986 + year_1987 + year_1988 + year_1989 + year_1990 + EntityEffects',
    data=fe_data
).fit()

print(fe_model.summary)
```

``Interpretation: Coefficient for lt is 0.78, meaning a 1% increase in temporary labor is associated with a 0.78% increase in output. Coefficient for k is 0.155, meaning a 1% increase in new capital is associated with a 0.155% increase in output.``

``Interpretation: Reject the null hypothesis that lambda_L = 1. Suggests that temporary labor is significantly less productive than permanent labor in the FE model as well.``

```python
# Extracting values
lambda_K = fe_model.params['k']
lambda_K_se = fe_model.std_errors['k']

# Calculating t-stat and p-value
t_stat_lambda_K = (lambda_K - 1) / lambda_K_se
p_value_lambda_K = 2 * (1 - norm.cdf(abs(t_stat_lambda_K)))

print('Test for lambda K = 1:')
print('lambda_K =', lambda_K)
print('SE =', lambda_K_se)
print('t-stat =', t_stat_lambda_K)
print('p_value =', p_value_lambda_K)
```

``Interpretation: Reject null hypothesis that lambda_K = 1. Suggests that new capital is significantly less productive than old capital in the FE model as well.``

```python
# Testing for aL + aK = 1
alpha_L = fe_model.params['lt']
alpha_K = fe_model.params['k']
alpha_sum = alpha_L + alpha_K

# Calculating SE, t-stat, p-value for the sum 
alpha_sum_se = (fe_model.cov.loc['lt', 'lt'] + fe_model.cov.loc['k', 'k'] + 2 * fe_model.cov.loc['lt', 'k']) ** 0.5
t_stat_alpha_sum = (alpha_sum - 1) / alpha_sum_se
p_value_alpha_sum = 2 * (1 - norm.cdf(abs(t_stat_alpha_sum)))

print('Test for aL + aK = 1:')
print('alpha_L=', alpha_L)
print('alpha_K', alpha_K)
print('t-stat =', t_stat_alpha_sum)
print('p_value =', p_value_alpha_sum)
```

``Interpretation: Reject null hypothesis that aL + aK = 1. Combined productivity of labor and capital less than 1 is significant evidence of decreasing returns to scale.``

#### (a) - Deriving the C-O Transformation

``Quasi-first transformation: yit - ρyi,t-1 = αL(lit - ρli,t-1) + αK(kit - ρki,t-1) + (year_dummiesit ρyear_dummiesi,t-1) + (1-ρ)ηi + εit``

``4 Restrictions: 1. Coefficient on lit / coefficient on Li,t-1 = -1/p 2. Coefficient on kit / coefficient on ki,t-1 = -1/p 3. Each year dummy coefficient / its lagged coefficient = -1/p 4. The term (1-p)ni captures the transformed fixed effect``


#### (b) - Estimating the Model Using Fixed Effects with Cochrane-Orcutt

```python
# Creating new dataset and setting the index for panel data structure 
co_data = data.copy()

co_data = pd.concat([data, year_dummies], axis=1)

co_data.set_index(['id', 'year'], inplace=True)

co_data
```

```python
# Getting residuals 
residuals = fe_model.resids

# Calculating logged residuals
co_data['residuals'] = residuals
co_data['lagged_residuals'] = co_data.groupby('id')['residuals'].shift(1)

co_data.info()
```

```python
import numpy as np

# Dropping missing values
filtered_data = co_data.dropna(subset=['residuals', 'lagged_residuals'])

# Regressing residuals on lagged residuals 
rho = np.sum(filtered_data['residuals'] * filtered_data['lagged_residuals']) / np.sum(filtered_data['lagged_residuals'] ** 2)

print('rho =', rho)
```

```python
co_data = co_data.sort_index()

# Applying quasi-first difference transformation
co_data['y_trans'] = co_data['y'] - rho * co_data.groupby('id')['y'].shift(1)
co_data['lt_trans'] = co_data['lt'] - rho * co_data.groupby('id')['lt'].shift(1)
co_data['k_trans'] = co_data['k'] - rho * co_data.groupby('id')['k'].shift(1)

co_data[['y_trans', 'lt_trans', 'k_trans']]
```

```python
# Creating lagged versions of the quasi-first difference transformed variables
co_data['y_trans_lag'] = co_data.groupby('id')['y_trans'].shift(1)
co_data['lt_trans_lag'] = co_data.groupby('id')['lt_trans'].shift(1)
co_data['k_trans_lag'] = co_data.groupby('id')['k_trans'].shift(1)

co_data[['y_trans_lag', 'lt_trans_lag', 'k_trans_lag']]

co_data.info()
```

```python
# Applying the quasi-first difference transformation on the existing year dummy variables
for year in range(1983, 1990):
    col = f'year_{year}'
    co_data[f'{col}_trans'] = co_data[col] - rho * co_data.groupby('id')[col].shift(1).fillna(0)

co_data[[f'year_{year}' for year in range(1983, 1990)] + [f'year_{year}_trans' for year in range(1983, 1990)]]
```

```python
co_data.columns
```

```python
co_data = co_data.dropna()
co_data.info()
```

```python
# Defining the model using the Fixed Effects C-O estimator with lagged varialbes and time dummies
fe_co_model = PanelOLS.from_formula(
    'y_trans ~ lt_trans + lt_trans_lag + k_trans + k_trans_lag + '
    'year_1983_trans + year_1984_trans + year_1985_trans + year_1986_trans + '
    'year_1987_trans + year_1988_trans + EntityEffects',
    data=co_data,
    check_rank=False,
    drop_absorbed=True
).fit()

# Display the summary
print(fe_co_model.summary)
```

``Interpretation: Coefficient for lt_diff is 0.59, meaning a 1 unit increase in the quasi-differenced temporary labor is associated with an significant increase of 0.59 in the quasi-differenced output. Coefficient for k_diff is 0.12, meaning a 1 unit increase in the quasi-differenced new capital is associated with an increase of 0.12 in the quasi-differenced output. Coefficients for the year dummy differences show both positive and negative shift in productivity as time goes by. ``

#### (c) - Testing the C-O Restrictions 

```python
fe_co_model.params.index
```

```python
from scipy.stats import chi2

# Testing Restriction 1: Coefficient ratio for labor
beta_lt = fe_co_model.params.get('lt_trans')
beta_lt_lag = fe_co_model.params.get('lt_trans_lag')

if beta_lt is not None and beta_lt_lag is not None:
    # Calculate the ratio
    lt_ratio = -beta_lt / beta_lt_lag

    # Calculate the standard error of the ratio
    lt_se = np.sqrt(
        (beta_lt_lag**2 * fe_co_model.std_errors['lt_trans']**2 + beta_lt**2 * fe_co_model.std_errors['lt_trans_lag']**2) / beta_lt_lag**4
    )    

    # Calculate the chi-squared test statistic
    chi2_lt = ((lt_ratio - (-1 / rho))**2) / lt_se**2
    chi2_lt_pvalue = 1 - chi2.cdf(chi2_lt, df=1)

    print('Test for 1st Restriction (lt):')
    print('Chi-squared statistic:', chi2_lt)
    print('p-value:', chi2_lt_pvalue)
```

```python
# Testing Restriction 2: Coefficient ratio for capital
beta_k = fe_co_model.params.get('k_trans')
beta_k_lag = fe_co_model.params.get('k_trans_lag')

if beta_k is not None and beta_k_lag is not None:
    # Calculating the ratio
    k_ratio = -beta_k / beta_k_lag

    # Calculating the standard error of the ratio
    k_se = np.sqrt(
        (beta_k_lag**2 * fe_co_model.std_errors['k_trans']**2 + beta_k**2 * fe_co_model.std_errors['k_trans_lag']**2) / beta_k_lag**4
    )

    # Calculating the chi-squared test statistic
    chi2_k = ((k_ratio - (-1 / rho))**2) / k_se**2
    chi2_k_pvalue = 1 - chi2.cdf(chi2_k, df=1)

    print('Test for 2nd Restriction (k):')
    print('Chi-squared statistic:', chi2_k)
    print('p-value:', chi2_k_pvalue)
```

```python
co_data.columns
```

```python
# Testing Restriction 3: Coefficient ratios for time dummies
for year in range(1984, 1990):
    current_param = next((param for param in fe_co_model.params.index if f'year_{year}_trans' in param), None)
    previous_param = next((param for param in fe_co_model.params.index if f'year_{year - 1}_trans' in param), None)
    
    if current_param and previous_param:
        beta_current = fe_co_model.params[current_param]
        beta_previous = fe_co_model.params[previous_param]
        se_current = fe_co_model.std_errors[current_param]
        se_previous = fe_co_model.std_errors[previous_param]
        
        # Calculating the ratio and the standard error
        year_ratio = -beta_current / beta_previous
        year_se = np.sqrt((beta_previous**2 * se_current**2 + beta_current**2 * se_previous**2) / beta_previous**4)
        
        # Calculating the chi-squared test statistic and p-value
        chi2_stat = ((year_ratio - (-1 / rho))**2) / year_se**2
        p_value = 1 - chi2.cdf(chi2_stat, df=1)
        
        # Output results
        print(f'Test for Year Ratio Restriction ({year - 1} to {year}):')
        print(f'  Chi-squared statistic: {chi2_stat:.4f}')
        print(f'  p-value: {p_value:.4f}\n')
```

#### (a) - Estimating the Parameters Using A-R Estimator

```python
# Creating new dataset and setting the index for panel data structure 
ar_data = data.copy()

ar_data = pd.concat([data, year_dummies], axis=1)

ar_data.set_index(['id', 'year'], inplace=True)

ar_data
```

```python
# Lagging dependent variable as an instrument
ar_data['y_lag1'] = ar_data.groupby('id')['y'].shift(1)

ar_data.info()
```

```python
ar_data = ar_data.dropna()
ar_data.info()
```

```python
# Defining the model using Arellano-Bond estimator with lagged y, lt, k, and time dummies
ab_model = PanelOLS.from_formula(
    'y ~ y_lag1 + lt + k + year_1983 + year_1984 + year_1985 + year_1986 + '
    'year_1987 + year_1988 + year_1989 + EntityEffects',
    data=ar_data
).fit()

print(ab_model.summary)
```

#### (b) - Testing for Serial Correlation in Transitory Shocks

```python
ar_data.info()
```

```python
# Creating new dataset
ar_residual_data = ar_data.copy()

# Getting residuals
residuals = ab_model.resids

# Calculating the lagged residuals
ar_residual_data.loc[:, 'residuals'] = residuals
ar_residual_data.loc[:, 'lagged_residuals'] = ar_residual_data.groupby('id')['residuals'].shift(1)

ar_residual_data.info()
```

```python
ar_residual_data = ar_residual_data.dropna()
ar_residual_data.info()
```

```python
# Calculating the correlation between residuals and lagged residuals
rho_residuals = np.corrcoef(ar_residual_data['residuals'], ar_residual_data['lagged_residuals'])[0, 1]
print("rho of Residuals = ", rho_residuals)
```

``Interpretation: rho of Residual is 0.022, indicating there is no significant evidence of serial correlation in transitory shocks.`` 

#### (c) - Over Identification Test

```python
ar_oit_data = ar_data.copy()

for year in range(1983, 1991):
    ar_oit_data[f'year_{year}'] = ar_oit_data[f'year_{year}'].astype(int)

ar_oit_data = ar_oit_data.reset_index()

ar_oit_data
```

```python
from linearmodels.iv import IV2SLS

# Defining dependent, exogenous, and instrumental variables explicitly
y = ar_oit_data['y']
X = ar_oit_data[['lt', 'k', 'year_1983', 'year_1984', 'year_1985', 'year_1986', 'year_1987', 'year_1988', 'year_1989']]
Z = ar_oit_data[['y_lag1']]


# Fit the model
iv_model = IV2SLS(y, X, None, Z).fit()

print(iv_model.summary)
```
